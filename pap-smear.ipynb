## Histogramas com classes destacadas por cores distintas

```python
# variável K/C em destaque

sns.displot(data=df, x='K/C', kind='hist',
            height=7, aspect=1.4, bins=20,
            hue='Class', hue_order=['1','2','3','4','5','6','7']);
```

```python
# demais variáveis

for feature in df.columns.values[1:-2]:
  sns.displot(data=df, x=feature, kind='hist',
            height=3, aspect=1.4, bins=10,
            hue='Class');
```

## Pairplot

```python
# somente variáveis com maior correlação com a coluna de classes

sns.pairplot(df, hue='Class', corner=True, 
             vars = ['Cyto_A','K/C','CytoShort','CytoLong','CytoRund','CytoPeri','CytoMax','CytoMin'])
```

## Mapeamento das classes para anormais e normais

```python
df_bin = df.copy()

mask = df_bin.Class <= 3
column_name = 'Class'
df_bin.loc[mask, column_name] = 0

mask = df_bin.Class > 0
column_name = 'Class'
df_bin.loc[mask, column_name] = 1

df_bin.hist('Class')
```

## Histogramas

```python
for feature in df_bin.columns.values[1:-2]:
  sns.displot(data=df_bin, x=feature, kind='hist',
            height=3, aspect=1.4, bins=10,
            hue='Class');
```

## Pairplot

```python
sns.pairplot(df_bin, hue='Class', kind='reg', corner=True, 
             vars = ['Cyto_A','K/C','CytoShort','CytoLong','CytoRund','CytoPeri','CytoMax','CytoMin'])
```

## Matriz de correlação - classes: normais e anormais

```python
matrix_corr_bin = df_bin.corr()
plt.figure(figsize=(20,15))
sns.heatmap(matrix_corr_bin, annot=True, cmap="Reds")
plt.show()
```
